```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Sample data (replace with your actual timing results)
threads = [1, 2, 4, 8]
sequential_time = 12.5  # Replace with your actual sequential time
parallel_times = [12.5, 6.8, 3.7, 2.3]  # Replace with your actual parallel times

# Calculate speedup
speedup = [sequential_time / t for t in parallel_times]
efficiency = [s/t for s, t in zip(speedup, threads)]

# Create dataframe
df = pd.DataFrame({
    'Threads': threads,
    'Time (s)': parallel_times,
    'Speedup': speedup,
    'Efficiency': efficiency
})

# Plotting
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.plot(df['Threads'], df['Time (s)'], 'bo-')
plt.xlabel('Number of Threads')
plt.ylabel('Execution Time (s)')
plt.title('Execution Time vs Threads')

plt.subplot(1, 3, 2)
plt.plot(df['Threads'], df['Speedup'], 'go-')
plt.xlabel('Number of Threads')
plt.ylabel('Speedup (x)')
plt.title('Speedup vs Threads')

plt.subplot(1, 3, 3)
plt.plot(df['Threads'], df['Efficiency'], 'ro-')
plt.xlabel('Number of Threads')
plt.ylabel('Efficiency')
plt.title('Efficiency vs Threads')

plt.tight_layout()
plt.show()

# Display table
display(df)
```
